```python
# we can see date and time columns are not date/time format

# the output columns are
# 1. Date Time (MM/DD/YYYY HH:MM:SS format)
# 2. Date (MM/DD/YYYY)
# 3. Time (HH:MM:SS format)
```

```python
import datetime
def object_to_datetime(x):
    date, hour = x[0], x[1].lower()
    check = [[0,1]["p" in hour], [0,1]["a" in hour]]
    hour = "".join(xx for xx in hour if xx.isdigit()) # remove character
    hour = str(int(hour) + check[0]*1200) # account for PM
    hour = "0"*(4-len(hour)) + hour # convert it to 4 lengths
    if check[1] and hour[:2] == "12":
        hour = "00" + hour[2:]
    
    #print(date+" "+hour)
    temp = datetime.datetime.strptime(date+" "+hour, "%m/%d/%y %H%M")
    return temp.strftime("%d/%m/%Y %H:%M:%S")

df["Date Time"] = df.apply(object_to_datetime, axis=1)
df["Date"] = df["Date Time"].apply(lambda x: x.split()[0])
df["Time"] = df["Date Time"].apply(lambda x: x.split()[1][:5].replace("00:", "0:"))
```

```python
df = df[["Date Time", "Date", "Time"]]
```
